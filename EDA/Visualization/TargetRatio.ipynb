```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data
train = pd.read_csv('../../Data/training_set.csv')
```

```python
# Make class ratio for plotting
num_0 = sum(train['class'] == 0)
num_1 = sum(train['class'] == 1)
ratio = [num_0, num_1]

# Assign plot obejct
fig, ax = plt.subplots(figsize = (20, 20))

# Plot pie graph
ax.pie(ratio, labels = ['0', '1'], autopct='%.1f%%',
       textprops = {'size': 20})

# Set legend
ax.legend(fontsize = 20)
```

```python
from sklearn.model_selection import train_test_split

X = train.drop('class', axis = 1, inplace = False).copy()
Y = train['class'].copy()

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size = 0.4, random_state = 31)
X_valid, X_test, Y_valid, Y_test = train_test_split(X_val, Y_val, test_size = 0.5, random_state = 31)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (10, 30))
for i, label in enumerate([Y_train, Y_valid, Y_test]):
    ratio = [sum(label == 0), sum(label == 1)]
    axes[i].pie(ratio, labels = ['0', '1'], autopct='%.1f%%')
    
plt.legend()
```
