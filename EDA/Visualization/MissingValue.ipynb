```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data
train = pd.read_csv('../../Data/training_set.csv')
```

```python
import missingno as msno
```

```python
# Make list which indicate position of bar
x_idx = [i for i in range(len(train.columns))]

# Assign plot object
fig, axes = plt.subplots(2, 1, figsize = (40, 20))

# Plot missingno form graph
msno.matrix(df = train, color = (0.1, 0.6, 0.8),
            ax = axes[0], sparkline = False)

# Plot bar object
sns.barplot(x = x_idx, y = train.isna().sum(),
            ax = axes[1])

# Set tick of axis
axes[1].set_xticks(x_idx)
axes[1].set_xticklabels(train.columns,
                        rotation = 90,
                        ha = 'center',
                        fontsize = 10)
yticks = axes[1].get_yticks()
axes[1].set_yticks(yticks)
axes[1].set_yticklabels([format(int(i), ',') for i in yticks],
                        fontsize = 30)

# Set grid
axes[1].grid(axis = 'y')
axes[1].set_axisbelow(True)
```

```python
# Assign column name list if needed
# cols = train.columns
cols = train.columns[40: 50]

# Assign plot object
fig, ax = plt.subplots(figsize = (20, 10))

# Plot missingno form graph
msno.matrix(df = train[cols], color = (0.1, 0.6, 0.8),
            ax = ax, sparkline = False)

# Set tick of axis
ax.set_xticks([i for i in range(len(cols))])
ax.set_xticklabels(cols,
                        rotation = 90,
                        ha = 'center',
                        fontsize = 20)

# Set Title
ax.set_title('MissingNo Graph',
             fontsize = 40)
```

```python
# Make list which indicate position of bar
x_idx = [i for i in range(len(train.columns))]

# Assign plot object
fig, ax = plt.subplots(figsize = (30, 10))

# Plot bar object
sns.barplot(x = x_idx, y = train.isna().sum(),
            ax = ax)

# Set tick of axis
ax.set_xticks(x_idx)
ax.set_xticklabels(train.columns,
                        rotation = 90,
                        ha = 'center',
                        fontsize = 10)
yticks = ax.get_yticks()
ax.set_yticks(yticks)
ax.set_yticklabels([format(int(i), ',') for i in yticks],
                        fontsize = 30)

# Set grid
ax.grid(axis = 'y')
ax.set_axisbelow(True)

# Set Title
ax.set_title('Number of MissingValue by each column',
             fontsize = 30)
```
